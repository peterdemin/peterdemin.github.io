```python
import PIL.Image
```

```python
inv = PIL.Image.open('/Users/peterdemin/Downloads/P005.png')
```

```python
import PIL.ImageOps    
```

```python
r,g,b,a = inv.split()
rgb_image = PIL.Image.merge('RGB', (r,g,b))
image = PIL.ImageOps.invert(rgb_image)
image
```

```python
cropped = image.crop((40, 380, 960, 620))
cropped.size
```

```python
import PIL.ImageDraw
```

```python
img = PIL.ImageDraw.Draw(cropped)
step = 920//5
for x in range(step, 920, step):
    img.rectangle([(x-10, 0), (x+10, 240)], fill ="#000000")
img.rectangle([(0, 110), (920, 130)], fill ="#000000")
cropped
```
